**Задание 1.** ***Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.***

Условию удовлетворяют два варианта- выпадение "2" или "5". Вероятность будет вычисляться как отношение балгоприятных исходов к общему количеству:

$Р=\frac{2}{6}=0,33$ или 33%.

**Задание 2.** ***Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.***

Вероятность выпадения "2":
$P_{2} = \frac{1}{6}$;

Вероятность выпадения "5":
$P_{5} = \frac{1}{6}$.

Вероятность последовательности выпадения двойки, а затем пятерки:
$P = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}=0,028$ или 2,8%.

**Задание 3.** ***Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости.***

В данном условии отсутствует указание порядка выпадения результатов, которые нам необходимо соблюдать, это значит, что необходимо рассмотреть оба варианта- выпадение "2", а затем "5"; и выпадение "5", а затем "2". Вероятности выпадения каждой из этих ситуаций рассчитываются как:

$P_{\text{{2-5}}} = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$;

$P_{\text{{5-2}}} = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$.

Полная вероятность будет равна сумме вероятностей этих событий:

$P = \frac{1}{36} + \frac{1}{36} = \frac{1}{18}=0,06$ или 6%.

**Задание 4.** ***На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?***

Длина отрезка, на который приходится местоположение точки С, удовлетворяющее условию- (15+9) - 20 = 4 см.

Вероятность попадания точки в этот промежуток- P = 4/20 = 0,2.

**Задание 5.** ***Телефонный номер состоит из 7 цифр. Какова вероятность, что это номер 8882777?***

Вариант, который бы удовлетворял условиям единственный, следовательно необходимо определить общее количество возможных комбинаций цифр в номере- $P = \frac{1}{10^7}.$

**Задание 6.** ***Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?***

Количество вариантов двузначного числа без нулей- 9 * 8 = 72 (9 вариантов для первой цифры и 9-1 для второй). Вероятность угадать число с первого раза $P = \frac{1}{72}=0,014$ или 1,4%.
