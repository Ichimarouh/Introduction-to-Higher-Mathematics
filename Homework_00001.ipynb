### Задание 4. Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.


```python
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 20, 101)
```

```python
const_1 = 1

const_2 = 2

y_1 = np.cos(x*const_1)

y_2 = np.cos(x*const_2)
```

```python
plt.plot(x, y_1, color='red')
plt.plot(x, y_2, color='blue')
plt.title('График cos(k*x)')
plt.xlabel('Значения x')
plt.ylabel('Значения cos(k*x)')
#plt.legend('best')
plt.show()
```
