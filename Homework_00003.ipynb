```python
def len_vect(vector):
    return np.sqrt(vector[0]**2 + vector[1]**2 + vector[2]**2)

a = np.array([10, 10, 10])
b = np.array([0, 0, -10])

print(f'Длина вектора a- {len_vect(a)},\n длина вектора b- {len_vect(b)}.')
```

### Задание 1.3 

**Напишите код на Python, реализующий построение графиков:**

**1. окружности,**	

**2. эллипса,**

**3. гиперболы.**

```python
# Окружность

def circle(x, r, pos=True):
    y = np.sqrt(r**2 - x**2)
    return y if pos else -y

r = 500
n_points = 10000
points = np.linspace(-r, r, n_points)
y1 = [circle(x, r) for x in points]
y2 = [circle(x, r, False) for x in points]
```

```python
plt.figure(figsize = (5, 5))    
plt.plot(points, y1, 'blue')  
plt.plot(points, y2, 'blue')
plt.axis('scaled')
plt.show()
```

```python
# Эллипс

def ellips(x, a, b, pos=True):
    y = np.sqrt(b**2 - b**2 * x**2 / a**2)
    return y if pos else -y


a, b = 30, 10
n_points = 10000
points = np.linspace(-a, a, n_points)
y1 = [ellips(x, a, b) for x in points]
y2 = [ellips(x, a, b, False) for x in points]
```

```python
plt.plot(points, y1, 'green')  
plt.plot(points, y2, 'green')
plt.show()
```

```python
# Гипербола

def hyperbola_hor(x, a, b, pos=True):
    y = np.sqrt(b**2 * x**2 / a**2 - b**2)
    return y if pos else -y

def hyperbola_vert(x, a, b, pos=True):
    y = np.sqrt(b**2 + b**2 * x**2 / a**2)
    return y if pos else -y


a, b = 40, 15
n_points = 10000
points = np.linspace(-3*a, 3*a, n_points)
y1 = [hyperbola_hor(x, a, b) for x in points]
y2 = [hyperbola_hor(x, a, b, False) for x in points]
```

```python
y1 = [hyperbola_vert(x, a, b) for x in points]
y2 = [hyperbola_vert(x, a, b, False) for x in points]

plt.plot(points, y1, 'orange')  
plt.plot(points, y2, 'orange')
plt.show()
```

**1. Нарисуйте трехмерный график двух параллельных плоскостей.**

**2. Нарисуйте трехмерный график двух любых поверхностей второго порядка.**

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Трехмерный график двух параллельных плоскостей

fig = figure(figsize = (10, 10))
ax = Axes3D(fig)

start, end = -20, 20

X = np.arange(start, end, 2)
Y = np.arange(start, end, 2)
X, Y = np.meshgrid(X, Y)

A, B, C, D1, D2 = -4, 8.2, 9, 105, -14

Z1 = (-A * X - B * Y - D1) / C
Z2 = (-A * X - B * Y - D2) / C

ax.plot_wireframe(X, Y, Z1, label='Ax + By + Cz + D1 = 0', linestyle='-.', linewidth=1)
ax.plot_wireframe(X, Y, Z2, colors='orange',label='Ax + By + Cz + D2 = 0')
plt.legend(frameon=False)
show()
```

```python
# Трехмерный график двух поверхностей второго порядка

fig = figure(figsize = (8, 8))
ax = Axes3D(fig)

a1, b1, a2, b2 = 1, 2, 2, 1

start, end = -20, 20

X = np.arange(start, end, 2)
Y = np.arange(start, end, 2)
X, Y = np.meshgrid(X, Y)

f = lambda x, y: x ** 2 / a1 ** 2 - y ** 2 / b1 ** 2

Z1 = f(X, Y)
Z2 = X**2 / a2**2 + Y**2 / b2**2 

ax.plot_surface(X, Y, Z1,  linewidth=1, color='blue')
ax.plot_surface(X, Y, Z2,  linewidth=1, color='orange')
show()
```

### Задание 2.2 
**Докажите, что при ортогональном преобразовании сохраняется расстояние между точками**

Примем расстояние между точками за длину вектора вида **|xy|** с координатами $ (x_1, y_1), (x_2, y_2) $

Преобразование в общем виде будет выглядеть: $ X = a_{11}*x + a_{12} * y + a_{13}, Y = a_{21} * x + a_{22} * y + a_{23} $

$ (X_2 - X_1) = a_{11}*x_2 + a_{12}*y_2 + a_{13} - (a_{11}*x_1 + a_{12}*y_1 + a_{13}) = a_{11}*(x_2 - x_1) + a_{12}*(y_2 - y_1) $, для Y преобразование аналогично.

Тогда длина преобразованного вектора **|XY|** (для простоты возьмем квадрат длины):

$ |XY|^2 = (X_2 - X_1)^2 + (Y_2 - Y_1)^2 = (a_{11}*(x_2 - x_1) + a_{12}*(y_2 - y_1))^2 + (a_{21}*(x_2 - x_1) + a_{22}*(y_2 - y_1))^2 $

После раскрытия скобок и объединения множителей получим:

$ |XY|^2 = (a_{11}^2 + a_{21}^2)*(x_2 - x_1)^2 + (a_{12}^2 + a_{22}^2)*(y_2 - y_1)^2 +2*(a_{11}*a_{12} + a_{21}*a_{22})*(x_2 - x_1)*(y_2 - y_1) $

Учитывая свойства ортогонального преобразования: $ a_{11}^2 + a_{21}^2 = 1 $,  $a_{12}^2 + a_{22}^2 = 1 $, $ a_{11}*a_{12} + a_{21}*a_{22} = 0 $, получим:

$ |XY|^2 = (x_2 - x_1)^2 + (y_2 - y_1)^2 = |xy|^2 $, т.е. **|XY| = |xy|**.

### Задание 2.4 

exp(x) + x∙(1 – y) = 1  
y = x2 – 1

Преобразуем систему, приведя к виду:

$ y = \frac{e^x + x - 1}{x} $

$ y = x^2 - 1 $

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-4, 4, 101)

plt.figure(figsize = (8, 12))
plt.plot(x, (x + np.exp(x) - 1) / x)
plt.plot(x, x**2 - 1)
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(-1, 5) 
plt.grid(True)
plt.axis('scaled')
plt.show()
```

```python
def equ_solve(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equ_solve, (2, 3))
x2, y2 = fsolve(equ_solve, (-2, -1))
              
print(x1, y1)
print(x2, y2)
```
